Dataset is highly unbalanced                                                               

*   0--> Normal  Transaction
*   1--> Fraudulent Transactions

```python
# separating the data for analysis
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

Undersampling


Build a sample dataset containing similar distribution of normal transactions and Fraud transactions

Number of Fraudulent Transaction -> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating 2 DataFrames

```python
#axis = 0 means row wise and axis = 1 means column wise concat
new_dataset = pd.concat([legit_sample, fraud], axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Feaures and Tragets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split Data into Training Data and Testing data

```python
X_train, X_test,Y_train,Y_test = train_test_split( X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Model Training

```python
# LogisticRegression
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train,Y_train)
```

Evaluvation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Training data accuracy : ",training_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Testing data accuracy : ",testing_data_accuracy)
```
